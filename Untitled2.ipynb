```
import numpy as np
import pandas as pd
def time_trimmer(time, voltage, decay_start):
    global index
    if decay_start == 0:
        for i in range(len(time)):
            if time[i] == 0:
                index = i
    else:
        for i in range(len(time)):
            if np.floor(time[i]) == decay_start:
                index = i

    # Correcting the time so that it starts to count when the meas starts
    corrected_time = time[index:]  # Creating a new array that goes from index to end
    corrected_time = corrected_time.reset_index()  # Resetting the index so that the series starts at 0
    corrected_time = corrected_time["second"]  # Renaming the new series
    # Saving the first value of the new, corrected series so that it does not change in the loop
    first_value = corrected_time[0]
    # Subtracting the first value in our corrected value to each value in the array to displace
    # everything to the left and make the plot start plotting at 0
    for i in range(len(corrected_time)):
        corrected_time[i] = corrected_time[i] - first_value

    # Correcting the voltage so that it starts when the meas starts
    corrected_voltage = voltage[index:]  # Creating a new array that goes from index (the point where the meas starts)
    # to the end of the array
    corrected_voltage = corrected_voltage.reset_index()  # Resetting the index so that the series starts at 0
    corrected_voltage = corrected_voltage["Volt"]  # Renaming the new series

    # Returns the corrected series
    return corrected_time, corrected_voltage


def period_finder(time, voltage, strip_no=1):
    # Allocate memory for all the series, the values are placeholders but
    # should be big enough for all the files
    strip_passing_oscilloscope = pd.Series(dtype=float, index=(i for i in range(30000)))
    periods = pd.Series(dtype=float, index=(i for i in range(10000)))
    v_ang = pd.Series(dtype=float, index=(i for i in range(10000)))
    corrected_time_values = pd.Series(dtype=float, index=(i for i in range(10000)))
    # loop counter
    j = 0
    for i in range(len(voltage)):  # Counts every voltage drop, which corresponds to the strip passing the oscilloscope
        if voltage[i] < 10:
            strip_passing_oscilloscope[j] = time[i]
            j = j + 1  # Increase the value of j so that the next value is placed in the next position in the array
    # Reset j to reuse it
    j = 0
    # Drop all the values in the strip_passing_oscilloscope series that have no values stored
    strip_passing_oscilloscope = strip_passing_oscilloscope.dropna()

    for i in range(len(strip_passing_oscilloscope) - 1):  # Finding the period of a revolution
        t1 = time[i]  # One value in the array
        t2 = strip_passing_oscilloscope[i + 1]  # Next value in the array
        periods[j] = t2 - t1  # The period is the difference between two consecutive times, since there's only 1 strip
        j = j + 1  # Increase the value of j so that the next value is placed in the next position in the array

    # Reset j to reuse it
    j = 0

    # Drop all the values in the times series that have no values stored
    periods = periods.dropna()

    for i in periods:  # Finding the angular velocities
        v_ang[j] = 2 * np.pi / (i * strip_no)  # w = 2 * PI / T, where T is the period
        j = j + 1

    # Drop all the values in the angular velocities series that have no values stored
    v_ang = v_ang.dropna()
    # v_ang = v_ang.reset_index()
    for i in range(len(v_ang)):
        if v_ang[i] > 250:
            v_ang[i] = 220
    # for i in range(len(strip_passing_oscilloscope)):
    #     strip_passing_oscilloscope[i] = strip_passing_oscilloscope[i] - strip_passing_oscilloscope[0]
    return v_ang, strip_passing_oscilloscope
```

```
import pandas as pd
df = pd.read_csv("/br047b.csv")
time = df["second"]
voltage = df["Volt"]
decay_start = 5
temp_time, temp_voltage = time_trimmer(time, voltage, decay_start)
v_ang, time_values = period_finder(temp_time, temp_voltage)


# df = pd.read_csv("/content/project_data_edited.csv")
df = pd.concat([v_ang, time_values], axis = 1).reset_index()
df.drop(columns=["index"], inplace = True)
df.columns = (["Omega", "Time"])
df.head(4)
```

```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib.ticker as ticker
%config InlineBackend.figure_format = 'retina'
from scipy.optimize import curve_fit
import random
 
#Time is our constant array
time = df['Time']
mass = 0.53035
 
'''define physical constants in S.I'''
electrical_conductivity = 1.59e7# of the metal
air_permitivity =  12.568e-7
coil_turns = 250                 #Fix me with a calculator 
disk_thickness = 2e-3
pole_area = 2.828e-3 
pole_disk_distance = 70e-3
radius= 15e-2
relative_permitivity = 1  #realtive permitittivity of aluminum apparently between 1.6-1.8 (source: http://www.mit.edu/~6.777/matprops/aluminum.html)
width_B_area = 1e-3
length_B_area = 4e-2
 
air_gap = 1e-2 #Isabel said it was positioned by eye and was around 1cm (1e-2 m)
 
'''Constants defined in the theory written by ignacio- start with "guesses and work out what they should be for this setup'''
 
model_hyperparams = {
        'N': 1,
        'zeta': 0.014,
        'C': 10,
        'd': -1
        }
 
model_weightings = {
    1: 1,
    2: 1,
    3: 1,
    4: 1
}
 
current = 0.048
```

```
'''Simplifying constants for the plug and chug'''
 
r = 1 / (electrical_conductivity * disk_thickness)
alpha = width_B_area / (electrical_conductivity * disk_thickness * length_B_area)
L = pole_disk_distance
area = (width_B_area * length_B_area) 
moment_intertia= (0.5* radius * mass)
 
m = (electrical_conductivity * area * disk_thickness * (pole_disk_distance**2) * alpha * moment_intertia)



 
def Tao(current, model_hparams: dict):
  '''
  Calc tao to find the magnetic braking ODE
  params~ mag field strength, time constant and constant m
  returns~ single value 
  '''
  B = (air_permitivity * relative_permitivity) * (coil_turns * current) / air_gap
 
  tao =  (model_hparams['zeta'] + model_hparams['C']*(B**2)) / (moment_intertia)
  return tao
 
def omega_braking(omega_initial, tao, current, model_hparams: dict):
  '''
  main system equasion for finding omega as a function of t,
  params ~ tao, time(array), current
  returns~ omega(array)
  '''
  B = (air_permitivity * relative_permitivity) * (coil_turns * current) / air_gap
  omega_initial_star = omega_initial + (model_hparams['N'] / (model_hparams['zeta'] + model_hparams['C']*(B**2)))
 
  omega = omega_initial_star * (np.exp(-time * tao)) - model_hyperparams['d']
  return omega
 
def plot_results(time, omega, omega_error, experimental_omega):
  fig = plt.figure(figsize=(10,6))
 
  ax = fig.add_subplot(111)
 
  # This nest bit does a lot of work and plots a graph with error bars.
  ax.errorbar(time,           # x coordinates
              omega,              # y coordinates
              marker='x',             # marker used is a circle 'o'. Could be crosses 'x', or squares 's', or 'none'
              markersize = 5,        # marker size
              color='black',          # overall colour I think
              ecolor='black',         # edge colour for you marker
              markerfacecolor='black',
              linestyle='none',       # no line joining markers, could be a line '-', or a dashed line '--'
              capsize=6,
              label = "Computationally obtained data"# width of the end bit of the error bars, not too small nor too big please.
              )
 
  ax.fill_between(time, omega + omega_error, omega- omega_error , alpha = 0.3, facecolor = 'blue', label = "error")
 
  #Plot the experimental data also 
 
  ax.errorbar(time,           # x coordinates
               experimental_omega,              # y coordinates
               marker='x',             # marker used is a circle 'o'. Could be crosses 'x', or squares 's', or 'none'
               markersize = 5,        # marker size
               color='green',          # overall colour I think
               ecolor='green',         # edge colour for you marker
               markerfacecolor='green',
               linestyle='none',       # no line joining markers, could be a line '-', or a dashed line '--'
               capsize=6,
               label = "Experimentally obtained data"
                             # width of the end bit of the error bars, not too small nor too big please.
               )
 
  plt.xlabel('Time (s)')       
  plt.ylabel('Angular velocity (m/s)')
 
  plt.tick_params(direction='in',      # I like 'in', could be 'out' or both 'inout'
                  length=12,            # A reasonable length
                  bottom='on',         
                  left='on',
                  top='on',
                  right='on',
                )
  plt.rcParams.update({'font.size':15})   # A decent font size so the text is legible. 
  plt.legend()
  plt.show()
 
def write_to_pd(time: list, omega_simulated: list):
  '''takes in a time numpy array and a list of simulated times into a df'''
 
  df_1 = pd.DataFrame([time, omega_simulated])
  df = df_1.T
  df.columns= ['time', 'omega']
  return df 
 
def concat_real_simulation(sim_df, experimental_df, time):
  '''takes in two dataframe objects, concats them, and finds the RMSE'''
  time_df = pd.DataFrame(time, columns=['time'])
 
  df = pd.concat([sim_df, experimental_df, time], axis=1)
  df.columns = ['simuation_omega', 'experimental_omega', 'time']
  rmse = (((sim_df.values.flatten() - experimental_df.values.flatten())**2).mean())**0.5
  
  df['error'] = (df['simuation_omega'] -df['experimental_omega'])
  
 
  mpe = (df['error']/100).mean()
 
  print("M.P.E:  ", mpe)
  print("RMSE: ", rmse)
  print("DF :", df)
  return df, rmse 
 
def rmse(sim_df, experimental_df, time):
  '''takes in two dataframe objects, concats them, and finds the RMSE'''
  time_df = pd.DataFrame(time, columns=['time'])
 
  df = pd.concat([sim_df, experimental_df, time], axis=1)
  df.columns = ['simuation_omega', 'experimental_omega', 'time']
  rmse = (((sim_df.values.flatten() - experimental_df.values.flatten())**2).mean())**0.5
  
  return rmse
```
